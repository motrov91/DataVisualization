#Aula 01

##primeras visualizaciones

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt

#Indicamos a nuestro notebook que genere graficos
%matplotlib inline

df = pd.read_csv('base_tiempo.csv')
df.head()
```

```python
#Transformamos la columna fecha en un formato de datatime
df['date'] = pd.to_datetime(df['fecha'])
df.info()
```

```python
#Dentro de los parentesis se agrega el (eje_X, eje_Y)
plt.plot(df['date'], df['temperatura'])
```

```python
#Cambiar configuracion del grafico
plt.figure(figsize=(18,8))
plt.plot(df['date'], df['temperatura'])
```

```python
#Agregando titulos
plt.figure(figsize=(18,8))
plt.plot(df['date'], df['temperatura'])
plt.title('Temperatura en el tiempo')
```

```python
#Generando los ejes
#Generando ejes puedo personalizar cada eje como yo desee
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
```

```python
#personalizar los ejes
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
eje.plot(df['date'], df['temperatura'])
eje.set_title('Temperatura en el tiempo')
```

```python
#Ubicando nombres a los ejes
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
eje.plot(df['date'], df['temperatura'])
eje.set_title('Temperatura en el tiempo')
eje.set_ylabel('Temperatura')
eje.set_xlabel('Date')
```

```python
#cambiando tamaños de las fuentes
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
eje.plot(df['date'], df['temperatura'])
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('Temperatura', fontsize=20)
eje.set_xlabel('Date', fontsize=20)
```

```python
#Adicionando leyendas
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
eje.plot(df['date'], df['temperatura'])
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('Temperatura', fontsize=20)
eje.set_xlabel('Date', fontsize=20)
eje.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

```python
#Cambiando el color del grafico
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
#cambiando color
eje.plot(df['date'], df['temperatura'], color='green')
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('Temperatura', fontsize=20)
eje.set_xlabel('Date', fontsize=20)
eje.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

#Aula 02

##Personalizando nuestras visualizaciones

```python
#Cambiando el color del grafico
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
#cambiando color
#lw cambiando el grosor de la linea
eje.plot(df['date'], df['temperatura'], color='g', lw = 4)
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('Temperatura', fontsize=20)
eje.set_xlabel('Date', fontsize=20)
eje.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

```python
#Cambiando el color del grafico
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
#cambiando color
#lw cambiando el grosor de la linea
eje.plot(df['date'], df['temperatura'], color='g', lw = 0.4)
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('Temperatura', fontsize=20)
eje.set_xlabel('Date', fontsize=20)
eje.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

```python
# Cambiando el estilo de la linea
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
#cambiando color
#ls estilo de la linea
eje.plot(df['date'], df['temperatura'], color='g', ls='dotted')
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('Temperatura', fontsize=20)
eje.set_xlabel('Date', fontsize=20)
eje.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

```python
# Trabajando con intervalores de tiempo
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
#cambiando color
eje.plot(df['date'], df['temperatura'], color='g')
#agregando intervalo de valores
#pasamos el contenido de xlim(en el formato que este el eje en este caso datetime)
eje.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('Temperatura', fontsize=20)
eje.set_xlabel('Date', fontsize=20)
eje.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

```python
# Ubicando marcadores
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
#marcadores
eje.plot(df['date'], df['temperatura'], color='g', marker='o')
#agregando intervalo de valores
#pasamos el contenido de xlim(en el formato que este el eje en este caso datetime)
eje.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('Temperatura', fontsize=20)
eje.set_xlabel('Date', fontsize=20)
eje.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

```python
# Ubicando una cuadricula
fig = plt.figure(figsize=(18,8))
eje = fig.add_axes([0, 0, 1, 1])
eje.plot(df['date'], df['temperatura'], color='g')
eje.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('Temperatura', fontsize=20)
eje.set_xlabel('Date', fontsize=20)
eje.legend(['Temperatura'], loc = 'lower right', fontsize=15)
#agregando cuadricula
eje.grid(True)
```
